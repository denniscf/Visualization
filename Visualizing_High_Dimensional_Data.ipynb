```python
def DisplayNDigits(digits, n, visualize):
    [digitsLbl, x, y] = digits.images.shape
    uniqueDigits = np.unique(digits.target).size
    sortedIdxs = np.argsort(digits.target)

    nDigitsToShow = n
    fig = plt.figure()
    for digit in range(0, uniqueDigits):
        count = 1
        for pos in range(0, sortedIdxs.size):
            if digit == digits.target[sortedIdxs[pos]]:
                for k in range(0, nDigitsToShow):
                    ax = fig.add_subplot(uniqueDigits/2, nDigitsToShow * 2, nDigitsToShow*digit + count)
                    ax.imshow(digits.data[sortedIdxs[pos + count - 1]].reshape(x, y), cmap=plt.cm.gray_r)
                    count += 1
            if count - 1 == nDigitsToShow:
                if digit != 9:
                    break
                else:
                    fig.suptitle('DIGITS DATASET - 2D MATRIX REPRESENTATION', fontsize=16)
                    plt.show()
                    return digits
```

```python
def LoadDigits(nDigitsPerPlotLine, visualize):
    digits = datasets.load_digits()
    if visualize:
        DisplayNDigits(digits, nDigitsPerPlotLine, visualize)
    return digits
```

```python
nDigitsPerPlotLine = 3
digits = LoadDigits(nDigitsPerPlotLine, True)
```

```python
######## DIMENSIONALITY REDUCTION ##########

# Depending on the number of dimensions and the number of points your dataset has, the size in memory of the matrix might be 
# simply too big for you to work with. A high enough number of dimensions (it is dependent on the algorithm, but normally in the 
# hundreds) might put you in the realm of "curse of dimensionality", which is characterized by algorithms scaling badly in respect
# to the number of dimensions and poor performance in respect to classification/regression. Examples of this are many, but a couple 
# of the most popular are KNN and clustering algorithms.
```

```python
# There are many techniques used to reduce the dimensionality of a dataset. Often this implies in selecting the most important 
# dimensions to represent the points in the dataset.

# In this training, we will see an example of PCA, another of Custom Feature Vectors and finally one of selecting the most relevant
# dimensions.
```

```python
######################## PCA - PRINCIPAL COMPONENT ANALYSIS ######################## 
```

```python
from sklearn.decomposition import PCA
from sklearn.lda import LDA
from mpl_toolkits.mplot3d import Axes3D
```

```python
def RunPCA(digits, dims, visualize):
    #Dimensionality Reduction
    pca = PCA(n_components=dims, whiten=True)
    dataPCA = pca.fit_transform(digits.data)
    pointSize = 50

    if visualize:
        fig = plt.figure()
        fig.suptitle('DIGITS AFTER PCA - {0}D'.format(dims), fontsize=16)
        if dims == 2:
            ax = fig.add_subplot(111)
            uniqueDigits = np.sort(np.unique(digits.target))
            graphHandles = []

            colors = cm.rainbow(np.linspace(0, 1, len(uniqueDigits)))
            for digit in uniqueDigits:
                idxs = np.where(digits.target == digit)[0]
                h = ax.scatter(dataPCA[idxs, 0], dataPCA[idxs, 1], c=colors[digit], s=pointSize*np.ones([1, idxs.size]), marker='o')
                graphHandles.append(h)
            plt.legend(graphHandles, uniqueDigits)
        elif dims == 3:
            ax = Axes3D(fig)
            uniqueDigits = np.sort(np.unique(digits.target))
            graphHandles = []

            colors = cm.rainbow(np.linspace(0, 1, len(uniqueDigits)))
            for digit in uniqueDigits:
                idxs = np.where(digits.target == digit)[0]
                h = ax.scatter(dataPCA[idxs, 0], dataPCA[idxs, 1], dataPCA[idxs, 2], c=colors[digit], s=pointSize*np.ones([1, idxs.size]), marker='o')
                graphHandles.append(h)
            #plt.legend(graphHandles, uniqueDigits)
        else:
            ax = fig.add_subplot(111)
            sortedIdxs = np.argsort(digits.target)
            im = ax.imshow(dataPCA[sortedIdxs,:], interpolation='nearest')
            fig.colorbar(im)
            ax.set_aspect('auto')
    plt.show()
    return dataPCA
```

```python
######################## LDA - LINEAR DISCRIMINANT ANALYSIS ######################## 
```

```python
def RunLDA(digits, dims, visualize):
    #Dimensionality Reduction
    lda = LDA(n_components=dims)
    dataLDA = lda.fit_transform(digits.data, digits.target)
    pointSize = 50

    if visualize:
        fig = plt.figure()
        fig.suptitle('DIGITS AFTER LDA - {0}D'.format(dims), fontsize=16)
        if dims == 2:
            ax = fig.add_subplot(111)
            uniqueDigits = np.sort(np.unique(digits.target))
            graphHandles = []

            colors = cm.rainbow(np.linspace(0, 1, len(uniqueDigits)))
            for digit in uniqueDigits:
                idxs = np.where(digits.target == digit)[0]
                h = ax.scatter(dataLDA[idxs, 0], dataLDA[idxs, 1], c=colors[digit], s=pointSize*np.ones([1, idxs.size]), marker='o')
                graphHandles.append(h)
            plt.legend(graphHandles, uniqueDigits)
        elif dims == 3:
            ax = Axes3D(fig)
            uniqueDigits = np.sort(np.unique(digits.target))
            graphHandles = []

            colors = cm.rainbow(np.linspace(0, 1, len(uniqueDigits)))
            for digit in uniqueDigits:
                idxs = np.where(digits.target == digit)[0]
                h = ax.scatter(dataLDA[idxs, 0], dataLDA[idxs, 1], dataLDA[idxs, 2], c=colors[digit], s=pointSize*np.ones([1, idxs.size]), marker='o')
                graphHandles.append(h)
            #plt.legend(graphHandles, uniqueDigits)
        else:
            ax = fig.add_subplot(111)
            sortedIdxs = np.argsort(digits.target)
            im = ax.imshow(dataLDA[sortedIdxs,:], interpolation='nearest')
            fig.colorbar(im)
            ax.set_aspect('auto')
    plt.show()
    return dataLDA
```

```python
def BuildFeatureVector(digits, normalize, visualization):
    [nDigits, nDims] = digits.data.shape
    featureVectors = np.zeros([nDigits, 3])
    for i in range(0, nDigits):
        allDims = digits.data[i, :]
        nValuePixelsTop = np.sum(allDims[0:nDims/2] > 0)
        nValuePixelsBottom = np.sum(allDims[nDims/2:] > 0)
        symmetry = np.std(abs(allDims[0:nDims/2] - allDims[nDims/2:]))

        featureVectors[i, 0] = nValuePixelsTop
        featureVectors[i, 1] = nValuePixelsBottom
        featureVectors[i, 2] = symmetry

    if normalize:
        featureVectors = (featureVectors - np.min(featureVectors, axis=0))/(np.max(featureVectors, axis=0) - np.min(featureVectors, axis=0))

    if visualization:
        pointSize = 50
        fig = plt.figure()
        fig.suptitle('FEATURE VECTOR - 3D', fontsize=16)
        ax = Axes3D(fig)
        uniqueDigits = np.sort(np.unique(digits.target))
        graphHandles = []

        colors = cm.rainbow(np.linspace(0, 1, len(uniqueDigits)))
        for digit in uniqueDigits:
            idxs = np.where(digits.target == digit)[0]
            h = ax.scatter(featureVectors[idxs, 0], featureVectors[idxs, 1], featureVectors[idxs, 2], c=colors[digit], s=pointSize*np.ones([1, idxs.size]), marker='o')
            graphHandles.append(h)
        #plt.legend(graphHandles, uniqueDigits)
        plt.show()
```

```python
def DisplayAllImagesAsMatrix(digits):
    sortedIdxs = np.argsort(digits.target)
    sortedData = digits.data[sortedIdxs, :]

    fig = plt.figure()
    fig.suptitle('ALL DIGITS ORDERED - EACH PIXEL AS A DIMENSION', fontsize=16)
    ax = fig.add_subplot(111)
    im = ax.imshow(sortedData)
    ax.set_aspect('auto')
    fig.colorbar(im)
    plt.show()
    return
```

```python
def factor_scatter_matrix(df, factor, palette=None):    
    import matplotlib.colors
    import numpy as np
    from pandas.tools.plotting import scatter_matrix
    from scipy.stats import gaussian_kde

    if isinstance(factor, basestring):
        factor_name = factor #save off the name
        factor = df[factor] #extract column
        df = df.drop(factor_name,axis=1) # remove from df, so it 
        # doesn't get a row and col in the plot.

    classes = list(set(factor))

    if palette is None:
        palette = ['#e41a1c', '#377eb8', '#4eae4b', 
                   '#994fa1', '#ff8101', '#fdfc33', 
                   '#a8572c', '#f482be', '#999999']

    color_map = {'Iris-setosa': '#d6c369',
 'Iris-versicolor': '#2dd4c2',
 'Iris-virginica': '#f9f538'}

    if len(classes) > len(palette):
        raise ValueError('''Too many groups for the number of colors provided.
We only have {} colors in the palette, but you have {}
groups.'''.format(len(palette), len(classes)))

    colors = factor.apply(lambda group: color_map[group])
    axarr = scatter_matrix(df,figsize=(10,10),marker='o',c=colors,diagonal=None)


    for rc in xrange(len(df.columns)):
        for group in classes:
            y = df[factor == group].icol(rc).values
            gkde = gaussian_kde(y)
            ind = np.linspace(y.min(), y.max(), 1000)
            axarr[rc][rc].plot(ind, gkde.evaluate(ind),c=color_map[group])

    return axarr, color_map
```

```python
from pandas.tools.plotting import scatter_matrix
plt.figure()
factor_scatter_matrix(data, "Name")
plt.show()
```
